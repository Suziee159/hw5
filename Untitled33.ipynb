```python
import pandas as pd
import matplotlib.pyplot as plt

# Read in the dataset from the local file
data = pd.read_csv("covid_19_data.csv")

# Group the data by the observation date and sum the confirmed, deaths, and recovered cases
grouped_data = data.groupby('ObservationDate').sum()

# Plot the data as a multi-line graph
plt.plot(grouped_data.index, grouped_data['Confirmed'], label='Confirmed Cases')
plt.plot(grouped_data.index, grouped_data['Deaths'], label='Deaths')
plt.plot(grouped_data.index, grouped_data['Recovered'], label='Recovered Cases')

# Add a title and axis labels
plt.title("COVID-19 Virus Spread")
plt.xlabel("Date")
plt.ylabel("Number of Cases")

# Show the legend
plt.legend()

# Display the graph
plt.show()
```
